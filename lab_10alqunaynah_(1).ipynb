```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/asukul/DS301-f19/master/datasets/divorce/divorce-csv.csv")
```

```python
#get the features and target
x=df.drop(["Class"],axis=1)
y=df["Class"]
print(x.shape)
print(y.shape)
```

```python
#TRAIN TEST SPLIT
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=24)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
# creat an instance of the model
dt=DecisionTreeClassifier()
#train
dt.fit(x_train,y_train)
#do prediction
preds=dt.predict(x_test)
#calculate the accuracy score
score=accuracy_score(y_test,preds)
score
```

```python
#import stuff needed to visulaze
#used to export tree
from sklearn.tree import export_graphviz
#used to run program from a python code
from subprocess import check_call
#used to read and show image in python
from PIL import Image
```

```python
#export the decission tree
export_graphviz(dt,out_file="tree.dot",feature_names=x.columns,max_depth=5,class_names=["Potential_divorce","Healthy Marriage"])
```

```python
from sklearn.tree import plot_tree
plot_tree(dt,feature_names=df.columns)
plt.savefig("mytree.png")
```
